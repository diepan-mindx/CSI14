# import các thư viện

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
```

# Thêm dữ liệu + xử lý dữ liệu thô

```python
df = pd.read_csv("temperatures.csv")
df
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.info()
```

```python
# chia input/ output
f_input = df['Farenheit'].to_numpy()
c_output = df['Celcius'].to_numpy()
type(f_input), type(c_output)
```

# huấn luyện mô hình

```python
# tao random
keras.utils.set_random_seed(42)
```

```python
# khai bao mo hinh
model = keras.Sequential([
    keras.layers.Dense(units=10, input_shape=[1]),
    keras.layers.Dense(units=1) # layer output
])

model.compile(loss='mean_squared_error', optimizer='adam')
model.summary()
```

```python
# huan luyen mo hinh
model.fit(f_input, c_output, epochs=100)
```

# Dự đoán

```python
f_user_input = np.array([90])
# thuc te
c_output_real = round((f_user_input[0] - 32)/ 1.8, 2)
# du doan
c_output_pred = model.predict(f_user_input)
# --------------------------
print(f"Thực tế: {c_output_real} độ C")
print(f"Dự đoán: {c_output_pred[0][0]:.2f} độ C")
```
